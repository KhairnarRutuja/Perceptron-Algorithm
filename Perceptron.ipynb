## Perceptron
- The Fundamental Building Block of Artificial Neural network.
- A Neural Network with single Input and output Layer that is called as single layer Perceptron.
- Its create two regions to classified two point in the data.

### IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### LOAD DATA

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

### PLOTTING THE POINTS

```python
plt.figure(figsize=(5, 3), facecolor="white")
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', data=df)
plt.show()
```

- Here we can seen that the data is liner

###  DEFINE INDEPENDANT AND DEPENDANT FEATURES

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

### IMPORT PECEPTRON MODEL

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

### TRAIN MODEL

```python
p.fit(X,y)
```

### FIND COEFFICIENTS

```python
p.coef_
```

### FIND INTERCEPT

```python
p.intercept_
```

### PLOTTING TWO REGIONS

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plt.figure(figsize=(5, 3), facecolor="white")
plot_decision_regions(X.values, y.values, clf=p, legend=2)
```

- Here we can seen that perceptron create two regions 
Note:- We not do data preprocessing step soo thats why points are not classified but our Aim is to create two regions first positive region and second is negative region 
- But how can i understand which region is positive or which is negative
- soo here perceptron is act like a line 
- line equation is **Ax+By+C=0**
- when **Ax+By+C>0** we identify the region is positive region and
- when **Ax+By+C<0** we identify the region is negative region.
