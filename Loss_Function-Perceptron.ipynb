## Loss Function in Perceptron
- In Perceptron use hinge loss

### IMPORT LIBRARIES

```python
from sklearn.datasets import make_classification
import numpy as np
import matplotlib.pyplot as plt
```

### LOAD DATA

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=15)
```

### PLOTTING POINTS

```python
plt.figure(figsize=(7,4))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=40)
plt.show()
```

```python
def perceptron(X, y):
    # Initialize weights and bias
    w1 = w2 = b = 1
    # Learning rate
    lr = 0.1
    
    # Iterate through a fixed number of epochs (here, 1000)
    for j in range(1000):
        # Iterate through each data point
        for i in range(X.shape[0]):
            # Compute the dot product of weights and input features plus bias
            z = w1 * X[i][0] + w2 * X[i][1] + b

            # Check if the point is misclassified
            if z * y[i] < 0:
                # Update weights and bias using the Perceptron update rule
                w1 = w1 + lr * y[i] * X[i][0]
                w2 = w2 + lr * y[i] * X[i][1]
                b = b + lr * y[i]
            
    return w1, w2, b
```

```python
w1,w2,b = perceptron(X,y)
```

```python
m = -(w1/w2)
c = -(b/w2)

print(m,c)
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + c
```

### PLOTTING GRAPH

```python
plt.figure(figsize=(7,4))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y, cmap='winter', s=40)

# Define the regions
plt.fill_between(x_input, y_input, min(y_input), where=(y_input >= min(y_input)), color='blue', alpha=0.3)
plt.fill_between(x_input, y_input, 2, where=(y_input <= 2), color='green', alpha=0.3)

plt.ylim(-3, 2)
plt.show()
```

- Using loss function we get best line.
